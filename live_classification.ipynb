# Live Classification

The accuracy of live classification is significantly lower than using the landmarks extracted from images to classify the ASL letters. This live classification is used to illustrate the project and does not provide accurate predictions.

```python
import pickle
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from sklearn import svm
from sklearn.metrics import confusion_matrix
import cv2
import mediapipe as mp
from sklearn.model_selection import GridSearchCV
```

```python
model_dict = pickle.load(open('./model.p', 'rb')) #load the model
model = model_dict['model']

cap = cv2.VideoCapture(0) #video capture

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
hands = mp_hands.Hands(static_image_mode=True, min_detection_confidence=0.3)

labels_dict = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E', 5: 'F', 6: 'G', 7: 'H', 8: 'I', 9: 'J', 10: 'K', 11: 'L', 12: 'M', 13: 'N', 14: 'O', 15: 'P', 16: 'Q', 17: 'R', 18: 'S', 19: 'T', 20: 'U', 21: 'V', 22: 'W', 23: 'X', 24: 'Y', 25: 'Z'}
while True: #start video frame

    ret, frame = cap.read()
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    results = hands.process(img_rgb)
    
    if results.multi_hand_landmarks: #save coordinates from hands 
        coordinates = []
        landmarks = []
        
        # get the landmark coordinates
        for hand_landmarks in results.multi_hand_landmarks:
            for i in range(len(hand_landmarks.landmark)):
                x = hand_landmarks.landmark[i].x
                y = hand_landmarks.landmark[i].y
                z = hand_landmarks.landmark[i].z
                coordinates.append(x)
                coordinates.append(y)
                coordinates.append(z)
            
            # draw the landmarks on the frame
            for hand_landmarks in results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(
                    frame,  
                    hand_landmarks, 
                    mp_hands.HAND_CONNECTIONS,  
                    mp_drawing_styles.get_default_hand_landmarks_style(),
                    mp_drawing_styles.get_default_hand_connections_style())
                
                if len(coordinates) == 63: #only append data that has the right amount of landmarkers (3x21)
                        landmarks.append(coordinates) #landmarks, x,y,z-coorindates, for each coordinate 21 points
                        prediction = model.predict(landmarks)
                        predicted_character = labels_dict[int(prediction[0])]
                        cv2.putText(frame, str(predicted_character), (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 5),5)

    cv2.imshow('frame', frame)
    #cv2.waitKey(1)
    key = cv2.waitKey(1)
    if key == 27:  # terminate code by pressing esc
        cap.release()
        cv2.destroyAllWindows()


cap.release()
```
